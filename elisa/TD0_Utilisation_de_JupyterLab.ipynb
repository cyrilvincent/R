JupyterLab est un environnement de développement interactif qui vous permet d'écrire et d'exécuter du code, principalement en Python (mais ici en R), dans des ***notebooks*** - des documents qui combinent du texte explicatif et du code exécutable. C'est un outil idéal pour l'apprentissage car il vous permet d'expérimenter avec le code, de voir les résultats immédiatement et de documenter votre processus de réflexion, le tout dans une interface conviviale accessible *via* votre navigateur web, sans avoir besoin d'installer quoi que ce soit.

## Interface de JupyterLab

Lors de votre première connexion vous ferez face à l'interface suivante :



Elle est divisée en plusieurs parties :

- **Sur la gauche** vous avez une barre d'outils qui s'ouvre sur l'interface de navigation entre fichiers et dossiers. Vous pouvez sélectionner le TD à ouvrir dans cette interface-là. Vous pouvez faire disparaître le volet contenant les noms de fichiers en cliquant dessus.
- **Sur la droite** vous avez l'interface de document, qui affichera le contenu des TD une fois que vous les aurez ouvert. En haut de celle-ci vous observez une barre avec des boutons qui permettent de sauver, ajouter, couper, copier, coller, exécuter, arrêter, ou ré-exécuter une cellule.  
  
- **Tout en haut** vous avez une barre de menus qui permet d'interagir avec la plateforme JupyterLab et interagir avec le *notebook* (c'est le nom du document) que vous avez ouvert.  


### Les cellules

Pour commencer ceux-ci est un *notebook* est divisé en **cellules**. Ce sont des blocs qui peuvent être de **code** ou bien de **texte** (selon une notation qu'on appelle le ***markdown***).
Un notebook peut être constitué d'autant de cellules que nécessaires. Ces blocs sont indépendants.

Vous pouvez sélectionner une cellule en cliquant dessus. Un cadre gris apparaît autour du bloc et une barre bleue sur le côté indiquant que la cellule a été sélectionnée.



Toutes les cellules sont ***a priori* éditables**. En double-cliquant sur une cellule vous pouvez la modifier. Pour le texte, cela vous montrera la version brute du texte, c'est-à-dire non mise en forme.



Pour récupérer le formatage initial du texte il faut "exécuter" le bloc de texte. Pour cela vous pouvez:
- utiliser le raccourci clavier `Majuscule+Entrée` (la touche majuscule est celle avec une flèche vers le haut **`↑`** ou bien `Contrôle+Entrée`,
- sélectionner la cellule et cliquer sur bouton exécuter en forme de bouton "play":  dans la barre au-dessus du notebook ouvert,
- passer par le menu "Run" puis "Run Selected Cell"  
  

Pour exécuter un bloc de code vous procéderez de la même manière.

Vous pouvez naviguer d'une cellule à l'autre en utilisant les flèches du clavier.

#### Créer une nouvelle cellule

Pour créer un nouveau cellule plusieurs options s'offrent à vous :

- Cliquer sur le bouton "+" de la barre d'outils du notebook 
- Double-cliquer sur une cellule et cliquer sur insérer une cellule au-dessus/en-dessous ,
- Cliquer à la toute fin du notebook ,
- Utiliser le raccourci clavier `B` quand une cellule est active (mais pas en train d'être éditée).

Lorsque vous voulez exécuter (ou ré-exécuter) toutes les cellules d'un notebooks vous pouvez utiliser le bouton en forme de double flèche  


Dans ce cas les cellules sont **executées de haut en bas et l'ordre a donc à une importance**.  
<span style="color:darkred">**Si vous rencontrez un problème c'est la première chose à faire.**</span>

#### Changer le type d'une cellule

Par défaut, quand vous créez une cellule, elle sera de type "Code" comme indiqué dans la barre d'outils  


Pour pouvoir taper et exécuter du test il faut changer de type et le définir comme "Markdown"  


Si vous vous êtes trompés de type il faut cette fois-ci sélectionner "Code"  


#### Écrire du code

Pour écrire du code il faut créer une cellule de type code. Les fonctions à taper seront étudiées tout au long du semestre et c'est l'objet du cours donc je ne vais pas les détailler ici.

Seulement, une chose importante à retenir c'est que, dans une cellule de code, toute ligne qui commence par dièse (`#`) ne sera pas exécutée. On l'appelle un **commentaire**. Cela permet d'expliquer ce que le code fait.

Voir la cellule en dessous.

```r
# Ce code multiple 2 par 4
2 * 4
```

Vous remarquez qu'après exécution de votre cellule de code un numéro est apparu entre crochets, cela correspond au numéro d'exécution de votre cellule par rapport aux autres du notebook.

Si vous avez des erreurs, il est possible que vous ayez oublié d'exécuter les cellules dans l'ordre, n'hésitez alors pas à relancer l'ensemble de votre notebook.

#### Écrire du texte

Pour écrire du texte il faut le taper dans une cellule qui a été définie comme une cellule "Markdown". Le **Markdown** est un langage de balisage léger pour formater un document, facile à lire, et facile à écrire.

Vous pouvez définir des titres de différents niveaux en utilisant les dièses `#`, une ligne précédée d'un seul dièse donnera un titre de niveau 1, deux un titre de niveau 2, etc.

Par exemple `# Titre` donne :

# Titre

Ou encore `## Sous-titre` donne :

## Sous-titre

Pour changer de paragraphe il faut revenir deux fois à la ligne.

Pour mettre du texte en italique il faut le mettre entre 1 étoiles `*` : `*texte en italique*`, ce qui donne *texte en italique*

Pour mettre du texte en gras il faut le mettre entre 2 étoiles `**` : `**texte en gras**`, ce qui donne **texte en gras**


Les possibilités sont multiples il est assez facile d'insérer des images ou des tables :
Pour aller plus loin voici un [lien vers la documentation de jupyter notebook (en anglais).](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html)

## Récupérer votre travail

Sur DOORS les notebooks sont accessibles tant que le cours est accessible. Vos modifications sur vos notebooks resteront persistentes toute la durée du semestre. Seulement pour assurer la pérennité de votre travail il peut être intéressant de télécharger une version de votre travail.

Vous pouvez le faire de deux façons différentes :

1. sous forme de notebook Jupyter
2. sous forme de PDF.

### Sous forme de notebook Jupyter

Pour récupérer les notebooks Jupyter vous devez ouvrir le notebook qui vous intéresse puis aller dans le menu "File" et cliquer sur "Download".  



Cela téléchargera automatiquement le notebook qui est ouvert en utilisant le même nom, avec un nom de fichier terminant par `.ipynb` ce qui correspond au format de fichier des notebooks Jupyter.

Vous pourrez le relancer soit en installant JupyterLab sur votre machine (par exemple en suivant les instructions sur [jupyter.org](https://jupyter.org)) ou bien en utilisant le service Jupyter offerts à l'UGA par le GRICAD (voir [leur site](https://gricad-jupyter.univ-grenoble-alpes.fr)) ou bien encore par exemple en utilisant [Google Colaboratory](https://colab.research.google.com/), qui est un service de Google proposant d'héberger des notebooks.

### Sous forme de PDF

Si vous souhaitez exporter plutôt une version PDF de votre notebook, qui contiennent tout le code et les figures exécutées, alors vous devez dans le menu "File" cliquer sur "Save and Export Notebook As..." puis "PDF".



Jupyter va alors transformer votre notebook en PDF en exécutant toutes les cellules. Il est possible que le rendu ne soit pas très joli. Vérifiez bien le contenu de votre PDF pour vous assurer que c'est conforme à ce que vous souhaitiez.
